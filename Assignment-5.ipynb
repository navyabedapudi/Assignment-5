```python
#Import Libraries
import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

### All required functions are in the below cell

```python
#Visualize Function
def visualize(datasets):
    df = pd.read_csv(datasets)
    temp_df = df
    temp_dict = dict()
    # Writing Average, Maximum, Minimum for each feature in a dataset to call later
    for key in temp_df.keys():
        temp_dict[key] = list()
        temp_dict[key].append(temp_df[key].mean())
        temp_dict[key].append(temp_df[key].max())
        temp_dict[key].append(temp_df[key].min())
    final_df = pd.DataFrame.from_dict(temp_dict).T
    final_df.columns = ['Average','max','min']
    
    #Creating subplots for vizualization
    fig,ax = plt.subplots(1,3,figsize = [30,10])
    i = 0
    for col in final_df.keys():
        ax[i].plot(np.array(final_df.index),np.array(final_df[col]))
        ax[i].set_title(col)
        i += 1 
    plt.show(fig)
        
#Function to indentiy if there are any issues in the dataset    
def identify_issues(datasets):
    df = pd.read_csv(datasets)
    if df[df.keys()[0]].max() == 0 and df[df.keys()[20]].max() == 20:
        print('I doubt the integrity of this dataset \n')
    else:
        mins = np.array([df[key].min() for key in df.keys()])
        mins_sum = mins.sum()
        if mins_sum == 0:
            print('Why this is happening \n')
        else:
            print('Dataset looks good \n')
            
        
    
```

### Run the code Below

```python
#Load Datasets
dataset = sorted(glob.glob('data*.csv'))
for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
